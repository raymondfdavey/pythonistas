````python
# Lottery Ticket Extension

# Lottery Simulation 

# Building on your previous lottery functions, create a lottery simulation.
# This extension will involve generating multiple tickets, simulating draws, and analyzing the results.

# Part A: Multiple Ticket Simulation

# Create a function simulate_lottery(num_tickets) that does the following:
# 1. Generate a single winning ticket using your previous function.
# 2. Generate a specified number of player tickets (num_tickets).
# 3. Check each player ticket against the winning ticket and calculate the winnings for each ticket in that draw.
# 4. Calculate and return the total winnings for those tickets and a breakdown of wins in each category.

# Hints:
# - Use your previously created functions: gen_random(), generate_ticket(), and check_win().
# - You may want to use a dictionary to keep track of the number of wins in each category.
# - Remember to sum up the winnings from all tickets.

# Example usage:
# total_winnings, win_breakdown = simulate_lottery(10)
# print(f"Total winnings: £{total_winnings}")
# print("Win breakdown:")
# for prize, count in win_breakdown.items():
#     print(f"  £{prize} wins: {count}")

# Part B: Jackpot Wait Time Simulation

# Create a function simulate_until_jackpot(num_tickets) that does the following:
# 1. Generate a set number of player tickets (num_tickets) and keep them constant.
# 2. Repeatedly generate new winning numbers (simulating weekly draws) and check against the player tickets.
# 3. Keep track of how many draws (weeks) it takes before one of the player tickets hits the jackpot (£1,000,000 win).
# 4. Return the number of weeks it took to hit the jackpot.

# Hints:
# - Use a while loop that continues until a jackpot is found.
# - Generate new winning numbers in each iteration of the loop.
# - Be careful! This simulation might take a long time to run. Consider adding a maximum number of weeks.

# Example usage:
# weeks_to_jackpot = simulate_until_jackpot(1000)
# print(f"It took {weeks_to_jackpot} weeks for one of the 1000 tickets to hit the jackpot!")

# Bonus Challenge:
# Modify your simulation to run multiple times (e.g., 100 simulations) and calculate the average number
# of weeks needed to hit a jackpot with a fixed set of tickets. This will give you a more reliable estimate of the odds.

# Remember to test your code and handle any potential errors!

# functions from previous solution

def gen_random(min_val, max_val):
    """Geerates a random number between min_val and max_val inclusive"""
    r = max_val - min_val + 1
    x = random.randrange(r) + min_val
    return x

def check_duplicate(my_list, val):
    """Checks to see if val is present in my_list, Returns True if present"""
    x = 0
    while x < len(my_list):
        if my_list[x] == val:
            return True
        else:
            x += 1
    return False

def generate_ticket():
    """Generate a sorted list with 6 elements in the range 1 to 50 inclusive"""
    my_list = []
    while len(my_list) < 6:
        x = gen_random(1, 50)
        if check_duplicate(my_list, x) is False:
            my_list.append(x)
    # Sort the list...
    my_list.sort(reverse=False)
    return my_list

def check_ticket(my_ticket, winning_numbers):
    matches = 0
    for x in my_ticket:
        for y in winning_numbers:
            if x == y:
                matches += 1
    if matches == 2:
        return 1
    elif matches == 3:
        return 10
    elif matches == 4:
        return 50
    elif matches == 5:
        return 500
    elif matches == 6:
        return 1000000
    else:
        return 0
````

````python
# ANSWER PART A

def simulate_lottery(num_tickets):
    """
    Simulates a lottery draw with multiple tickets
    Returns total winnings and breakdown of wins by prize category
    """
    # Generate winning numbers
    winning_numbers = generate_ticket()
    
    # Initialize tracking variables
    total_winnings = 0
    win_breakdown = {0: 0, 1: 0, 10: 0, 50: 0, 500: 0, 1000000: 0}
    
    # Generate and check each ticket
    for i in range(num_tickets):
        ticket = generate_ticket()
        winnings = check_ticket(ticket, winning_numbers)
        total_winnings += winnings
        win_breakdown[winnings] += 1
    
    return total_winnings, win_breakdown

# Part A: Multiple Ticket Simulation
print("\nSimulating 1000 lottery tickets...")
total_winnings, win_breakdown = simulate_lottery(1000)
print(f"Total winnings: £{total_winnings:,}")
print("Win breakdown:")
for prize, count in win_breakdown.items():
    if prize == 0:
        print(f"  No wins: {count}")
    else:
        print(f"  £{prize:,} wins: {count}")
````

````python
# ANSWER PART B

def simulate_until_jackpot(num_tickets, max_weeks):  # 100 years maximum
    """
    Simulates weekly draws until one ticket hits the jackpot
    Returns number of weeks it took
    Includes a max_weeks parameter to prevent infinite loops
    """
    # Generate fixed set of player tickets
    player_tickets = [generate_ticket() for _ in range(num_tickets)]
    
    weeks = 0
    jackpot_found = False
    
    while not jackpot_found and weeks < max_weeks:
        weeks += 1
        winning_numbers = generate_ticket()
        
        # Check each ticket for jackpot
        for ticket in player_tickets:
            if check_ticket(ticket, winning_numbers) == 1000000:
                jackpot_found = True
                break
    
    if not jackpot_found:
        return None  # Indicate no jackpot was found within max_weeks
    
    return weeks

# Part B: Jackpot Wait Time Simulation
num_tickets = 1000
max_weeks=52*1000 # 1000 years
print(f"\nSimulating until jackpot with {num_tickets} tickets...")
weeks = simulate_until_jackpot(num_tickets, max_weeks)
if weeks is None:
    print(f"No jackpot found within {max_weeks} weeks time limit!")
else:
    print(f"It took {weeks:,} weeks for one of the 1000 tickets to hit the jackpot!")
````

````python
# ANSWER PART C
# will take a long time to run depending on max_weeks
def simulate_average_jackpot_time(num_tickets, max_weeks, num_simulations=100):
    """
    Runs multiple jackpot simulations and calculates the average weeks to win
    """
    total_weeks = 0
    successful_sims = 0
    
    for _ in range(num_simulations):
        result = simulate_until_jackpot(num_tickets, max_weeks)
        if result is not None:
            total_weeks += result
            successful_sims += 1
    
    if successful_sims == 0:
        return None
    
    return total_weeks / successful_sims


# Bonus: Average Jackpot Time
num_tickets = 1000
num_simulations = 10
max_weeks = 52*1000 # 1000 years
print("\nCalculating average jackpot time over 10 simulations...")
avg_weeks = simulate_average_jackpot_time(num_tickets, max_weeks, num_simulations)
if avg_weeks is None:
    print("No jackpots found in any simulation!")
else:
    print(f"Average weeks to hit jackpot with {num_tickets} tickets per week: {avg_weeks:,.1f}")
````

````python
# Word Guessing Game

 # In this challenge, you have to create a word guessing game similar to hangman.
# The rules are:
# • A word is selected at random from a predefined list.
# • The user then has a maximum number of attempts to guess the word letter by letter.
# • If the user guesses a correct letter, all instances of that letter in the word are revealed.
# • If the user guesses incorrectly, they lose an attempt.
# • If the user enters a letter they've already guessed, a message is shown and the attempt count is not increased.
# • The user can quit the game by entering '-1'.

# All user guesses should be stored in a list for convenience.

# To build this game, divide the functionality into these functions:

# • generate_word(): should return a randomly selected word from a predefined list.

# • analyse_guess(word, guess): should take the word and the user's guessed letter as parameters.
#   It should return True if the letter is in the word, False otherwise.

# • check_previous(guessed_letters, guess): should take the list of previous guesses and the current guess.
#   It should return True if the guess is already in the list, False otherwise.

# • update_word_state(word, word_state, guess): should take the word, current word state, and guessed letter.
#   It should return an updated word state with the guessed letter revealed if correct.

# Hints:
# - Use underscores to represent unrevealed letters in the word state.
# - current_word_state = ["_"] * len(word_to_guess) will create a string of underscored equal to the word to guess length
# - Remember to handle both uppercase and lowercase letter inputs.
# - You can use a list of words like: ["python", "programming", "computer", "algorithm", "function", "variable"]
# - The random.choice() function can be used to select a random word from the list.
# - You can use an empty return to exit a main() function or break to break out of a loop
# - Word state: This is a string representing the current revealed state of the word to be guessed.
# - For example, if the word is "python" and the player has guessed 'p' and 'n',
# - the word state would be "p____n".

# - Your code here...

````

````python
import random

def generate_word():
    """Returns a randomly selected word from a predefined list"""
    words = ["python", "programming", "computer", "algorithm", "function", "variable"]
    return random.choice(words)

def analyse_guess(word, guess):
    """Check if the guessed letter is in the word"""
    return guess.lower() in word.lower()

def check_previous(guessed_letters, guess):
    """Check if the letter has been guessed before"""
    return guess.lower() in [letter.lower() for letter in guessed_letters]

def update_word_state(word, word_state, guess):
    """Update the word state with correctly guessed letter"""
    guess = guess.lower()
    word = word.lower()
    new_state = list(word_state)
    
    for i, letter in enumerate(word):
        if letter == guess:
            new_state[i] = guess
            
    return "".join(new_state)

# Game setup
word_to_guess = generate_word()
attempts_left = 6
guessed_letters = []
current_word_state = "_" * len(word_to_guess)

print("Welcome to Word Guessing Game!")
print(f"The word has {len(word_to_guess)} letters: {current_word_state}")

# Main game loop
game_quit = False  # Add a flag to track if player quit

while attempts_left > 0 and "_" in current_word_state:
    print(f"\nAttempts left: {attempts_left}")
    print(f"Current word: {current_word_state}")
    print(f"Guessed letters: {', '.join(guessed_letters)}")
    
    # Get player's guess
    guess = input("Enter a letter (or -1 to quit): ").strip()
    
    # Check for quit condition
    if guess == "-1":
        print(f"\nGame ended. The word was: {word_to_guess}")
        game_quit = True  # Set the flag
        break
        
    # Validate input
    if len(guess) != 1 or not guess.isalpha():
        print("Please enter a single letter!")
        continue
        
    # Check if letter was already guessed
    if check_previous(guessed_letters, guess):
        print("You already guessed that letter!")
        continue
        
    # Add to guessed letters
    guessed_letters.append(guess.lower())
    
    # Check if guess is correct
    if analyse_guess(word_to_guess, guess):
        current_word_state = update_word_state(word_to_guess, current_word_state, guess)
        print("Correct guess!")
    else:
        attempts_left -= 1
        print("Wrong guess!")

# Game end - only show win/lose message if player didn't quit
if not game_quit:  # Only show these messages if player didn't quit
    if "_" not in current_word_state:
        print(f"\nCongratulations! You won! The word was: {word_to_guess}")
    else:
        print(f"\nGame Over! The word was: {word_to_guess}")
````

````python
#Answer:

class Customer:
    """A class to represent a customer for our cafe."""

    def __init__(self, name, loyalty_discount, allergies=None):
        self.name = name
        self.loyalty_discount = loyalty_discount
        self.allergies = allergies  # If no allergies passed in, it will be None
class Order:
    """A schedule of items ordered and the cost for a particular customer."""

    def __init__(self, customer):
        self.customer = customer
        self.items_ordered = []
        self.cost = 0.0

    def add_to_order(self, item):
        """Add an item to the order and update the cost."""
        self.items_ordered.append(item)
        self.cost += item.price

    def summarise_order(self):
        """Display a summary of the order, including any loyalty discount."""
        # Show customer name and allergies if they have any
        if self.customer.allergies:
            print(f'Customer: {self.customer.name} (Allergies: {", ".join(self.customer.allergies)})')
        else:
            print(f'Customer: {self.customer.name} (No Allergies Stated)') 
            
        print(f'Total items ordered: {len(self.items_ordered)}')
        for i in self.items_ordered:
            i.display_details()
            
        # Apply any loyalty discounts...
        if self.customer.loyalty_discount is True:
            savings = 0
            for i in self.items_ordered:
                savings += i.discount
            self.cost -= savings
            print(f'Today you saved £{savings:.2f}')
        print(f'Total cost: £{self.cost:.2f}')
        
        # Check and display allergen warnings
        if self.customer.allergies:
            allergen_warnings = []
            for item in self.items_ordered:
                if item.allergens:
                    for allergen in item.allergens:
                        if allergen in self.customer.allergies:
                            warning = f"Warning! {item.name} contains {allergen}, which customer is allergic to."
                            allergen_warnings.append(warning)
            
            if allergen_warnings:
                print("\nAllergen Warnings:")
                for warning in allergen_warnings:
                    print(f"- {warning}")

class Item:
    """A super class for all food and drink that can be ordered in the cafe."""

    def __init__(self, name, price, allergens=None):
        self.name = name
        self.price = price
        self.allergens = allergens  # If no allergens passed in, it will be None

class Drink(Item):
    def __init__(self, name, price, size, allergens=None):
        super().__init__(name, price, allergens)
        self.size = size
        self.discount = 0.10

class Tea(Drink):
    def __init__(self, name, price, size, flavour, allergens=None):
        super().__init__(name, price, size, allergens)
        self.flavour = flavour

    def display_details(self):
        """Display details of the tea, including the flavour."""
        print(f'{self.name} £{self.price:.2f}, {self.size}, {self.flavour}')

class MineralWater(Drink):
    def __init__(self, name, price, size, is_carbonated, allergens=None):
        super().__init__(name, price, size, allergens)
        self.is_carbonated = is_carbonated

    def display_details(self):
        """Display details of the water, including whether it is carbonated."""
        print(f'{self.name} £{self.price:.2f}, {self.size}')
        if self.is_carbonated:
            print('Sparking water')
        else:
            print('Still water')

class Cake(Item):
    def __init__(self, name, price, slice_size, kind, allergens=None):
        super().__init__(name, price, allergens)
        self.slice_size = slice_size
        self.kind = kind
        self.discount = 0.0

    def display_details(self):
        """Display details of the cake."""
        print(f'{self.name} £{self.price:.2f}, {self.slice_size}, {self.kind}')

class Sandwich(Item):
    def __init__(self, name, price, bread_type, filling, allergens=None):
        super().__init__(name, price, allergens)
        self.bread_type = bread_type
        self.filling = filling
        self.discount = 0.20

    def display_details(self):
        """Display details of the sandwich, including the type of bread and filling."""
        print(f'{self.name} £{self.price:.2f}, {self.bread_type}, {self.filling} filling')


def main():
    # Create customers with allergies...
    cust1 = Customer('Harry Palmer', False, ["nuts"])  
    cust2 = Customer('Bill Preston', True, ["lactose", "gluten"])  # A loyal regular customer
    cust3 = Customer('Sarah Connor', False)  # Customer with no allergies

    # Order some items...
    order1 = Order(cust1)
    order1.add_to_order(Tea('Black tea', 2.00, 'large', 'Earl Gray'))
    order1.add_to_order(Cake('Chocolate dream', 2.30, 'medium', 'chocolate', ["nuts"]))

    order2 = Order(cust2)
    order2.add_to_order(MineralWater('Evian', 1.50, 'small', False))
    order2.add_to_order(Sandwich('Simple sandwich', 1.50, 'white', 'cheese', ["gluten", "lactose"]))

    order3 = Order(cust3)
    order3.add_to_order(Cake('Nutty delight', 3.00, 'large', 'walnut', ["nuts"]))

    # Summarise our orders...
    order1.summarise_order()
    print()
    order2.summarise_order()
    print()
    order3.summarise_order()
    print()

if __name__ == "__main__":
    main()
````
